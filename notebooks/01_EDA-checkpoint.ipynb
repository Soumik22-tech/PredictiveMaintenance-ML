# 1. LOAD DATA
Import pandas, numpy, matplotlib.pyplot, and seaborn. Set the seaborn style to 'darkgrid'. Load the CSV from '../data/predictive_maintenance.csv' into a DataFrame named `df`. Display `df.shape`, `df.head()`, `df.dtypes`, and `df.describe()`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set plotting style
sns.set_theme(style="darkgrid")

# Create models directory if it doesn't exist
os.makedirs('../models', exist_ok=True)

# Load data
df = pd.read_csv('../data/predictive_maintenance.csv')

# Data inspection
print(f"Dataset Shape: {df.shape}")
print("\nFirst 5 rows:")
display(df.head())
print("\nData Types:")
print(df.dtypes)
print("\nStatistical Summary:")
display(df.describe())
```

# 2. MISSING VALUES & DUPLICATES
Visualize missing values using `sns.heatmap(df.isnull())`. Calculate and print the count of duplicate rows using `df.duplicated().sum()`. Drop any duplicates found using `df.drop_duplicates()`.

```python
# Missing values check
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# Duplicates check
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

if duplicate_count > 0:
    df = df.drop_duplicates()
    print("Dropped duplicates.")
```

# 3. TARGET VARIABLE ANALYSIS
Generate countplots for the 'Failure Type' (multiclass) and 'Machine failure' (binary) columns with `figsize=(12,5)`. Calculate the percentage distribution for 'Failure Type'. 

**Note:** This dataset is heavily imbalanced toward the "No Failure" class. This imbalance must be addressed during model training.

```python
# Target Distribution
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

sns.countplot(data=df, x='Machine failure', ax=ax[0])
ax[0].set_title('Distribution of Machine Failure (Binary)')

sns.countplot(data=df, y='Failure Type', ax=ax[1])
ax[1].set_title('Distribution of Failure Type (Multiclass)')

plt.tight_layout()
plt.savefig('../models/target_distribution.png')
plt.show()

# Class distribution percentage
print("Failure Type distribution percentage:")
print(df['Failure Type'].value_counts(normalize=True) * 100)
```

# 4. FEATURE DISTRIBUTIONS
Create a grid of histograms for 'Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', and 'Tool wear [min]' using a `(18,12)` figsize. Apply `plt.tight_layout()` and save the figure to '../models/feature_histograms.png'.

```python
# Feature Distributions
features = ['Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']

plt.figure(figsize=(18, 12))
for i, col in enumerate(features, 1):
    plt.subplot(3, 2, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.savefig('../models/feature_histograms.png')
plt.show()
```

# 5. CORRELATION ANALYSIS
Compute the correlation matrix for numerical features and visualize it using `sns.heatmap` with `annot=True`. Set `figsize=(12,5)`. Save the figure to '../models/correlation_heatmap.png'.

```python
# Correlation Matrix
corr_matrix = df[features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')

plt.tight_layout()
plt.savefig('../models/correlation_heatmap.png')
plt.show()
```

# 6. FEATURE vs TARGET (BOXPLOTS)
Analyze the relationship between numerical features and 'Failure Type' using a grid of boxplots with `figsize=(18,12)`. Ensure each plot has appropriate titles and labels. Save the figure to '../models/feature_target_boxplots.png'.

```python
# Feature vs Target Analysis
plt.figure(figsize=(18, 15))
for i, col in enumerate(features, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(data=df, x=col, y='Failure Type')
    plt.title(f'{col} by Failure Type')

plt.tight_layout()
plt.savefig('../models/feature_target_boxplots.png')
plt.show()
```

# 7. CLASS IMBALANCE SUMMARY
Print the exact value counts for each category in 'Failure Type' and calculate the imbalance ratio. 

**Note:** We will handle this imbalance using SMOTE (Synthetic Minority Over-sampling Technique) in the preprocessing stage to ensure our model generalizes well to minority failure classes.

```python
# Class Imbalance Summary
counts = df['Failure Type'].value_counts()
print("Value counts for Failure Type:")
print(counts)

majority = counts.max()
minority_min = counts.min()
imbalance_ratio = majority / minority_min
print(f"\nImbalance Ratio (Majority : Rarest Minority): {imbalance_ratio:.2f} : 1")
```
